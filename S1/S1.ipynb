Condicional IF

```python
lista_dos = [2,4,6,7]
pares = []
impares = []

for i in lista_dos:
  if i%2 == 0:
    pares.append(i)
  else:
    impares.append(i)

print((pares, impares))
print(pares,impares)
```

Ciclo While

```python
a = 10

while a < 15:
  a = a + 1
  print(a)
  
```

```python
## A cada elemento de la lista de impares, les vamos a sumar 5 si es que es impar y es menor o igual a 5, de lo contrario,
## se cierra el ciclo

impares = [1,3,5,7,8,9]
impares_sumados = []

for i in impares:
  print(i) 
  while i <= 5:
    if i%2 != 0: 
      a = i + 5 
      impares_sumados.append(a)
      break
    else:
      print('No se cumple la condición')
      break
print(impares_sumados)
```
