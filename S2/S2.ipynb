```python
#### Tarea:

### Hacer un programa que pida dos números. Si es que es una división entre cero, mostrar error "No se puede dividir entre cero". Si se ingresa un texto en cualquiera de los dos números
### el programa deberá pedir nuevamente el ingreso del número que está erróneo.


a = input("Ingrese el primer numero: ")
b = input("Ingrese el segundo numero: ")

while type(a) != int or type(b) != int:
  try:
    a = int(a)
    b = int(b)
    print("La division entre su primer numero y el segundo numero es ",a/b, " la division entre su segundo numero y el primero es ",b/a)
  except ZeroDivisionError:
    print("No se puede dividir para cero")
  except:
    
    if type(a) == int and type(b) != int:
      b = input("Ingrese nuevamente el segundo numero: ")
    elif type(b) == int and type(a) != int:
      a = input("Ingrese nuevamente el primer numero: ")

```
